```python

#*************************************************************************************************

#           Program  1 :  To Find the Function for Data Given and plot the fit

#*************************************************************************************************
#importing the packages and setting the alias

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
 
from scipy import interpolate
from scipy import optimize

# Data Given, The max and min temperatures monthwise for the year

xmonths = np.arange(12)
tempmax = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
tempmin = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18]) 

# finding the function that relates the max temperatures and months
# To Calculate the params and param_covariance by calling optimize curve fit function and giving offsets

def yearly_temps(times, avg, ampl, time_offset):
    return (avg + ampl * np.cos((times + time_offset) * 2 * np.pi / times.max()))


# for new x values within that duration range 
xdays = np.linspace(0,11,365)



res_max, cov_max = optimize.curve_fit(yearly_temps, xmonths, tempmax, [20, 10, 0])
res_min, cov_min = optimize.curve_fit(yearly_temps, xmonths, tempmin, [-40, 20, 0])

# The optimized curve fit is obtained by adding the residual  

# Plotting the Curved Fit 

plt.plot(xmonths,tempmax,'bo')
plt.plot(xdays,yearly_temps(xdays,*res_max),'b-')



plt.plot(xmonths,tempmin,'ro')
plt.plot(xdays,yearly_temps(xdays,*res_min),'r-')

# Setting the labels of x and y axis 
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')
plt.show()



#*************************************************************************************************
```

```python
#*************************************************************************************************

#           Program  2 :  To read the  CSV file 
#                         1. Create Pie Chart , presenting Male Female Proportion
#                         2. Create ScatterPlot with Fare paid and the Age , differ the plot color by gender

#*************************************************************************************************
# Importing the packages and setting an alais 

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sp


# reading the csv file and populating the DataFrame 

df = pd.read_csv('https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv')


# 1. To Create a Pie Chart of MAle and FEmale distribution , as the sex column is string , creating another column 
# of numbers 1 for female and 0 for male
#****************************************************************************************************************
df['sexInNum'] = np.where((df['sex'] == 'female') ,1 ,0) 
print("The Female and Male distribution is : \n",'-'*100,'\n',df.groupby('sex')['sexInNum'].count())

# Plotting the Gender Distribution
print("The Plot for Gender Distribution :\n",'-'*80)
plt.pie(df.groupby('sex')['sexInNum'].count(),labels=('Female','Male'))


#****************************************************************************************************************
# 2. To Create Scatter plot of Fare paid and Age , differ the plot color by gender
# as the matplotlib doesnt give categorywise colours ,so using seaborn lmplot which gives this feature

#****************************************************************************************************************

sp.lmplot(x = "fare",y = "age",data=df, hue="sex")
print("\n The Plot for Fare Paid and Age by Gender :\n",'-'*80)



#****************************************************************************************************************
```
